```python
import pandas as pd
import itertools


param_values = {'gamma': [0.9, 0.95, 0.99, 0.999],
                'learning_rate': [1e-4, 1e-3, 1e-2, 1e-2]}

def make_param_grid(param_values):
    keys=param_values.keys()
    combinations=itertools.product(*param_values.values())
    df = pd.DataFrame.from_dict([dict(zip(keys,cc)) for cc in combinations])
    df[' ']


make_param_grid(param_values)
```
